```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#2.1
planets_df = sns.load_dataset('planets')

plt.figure(figsize=(25, 10))
sns.scatterplot(data = planets_df, x = 'orbital_period', y = 'mass', hue = 'method')
plt.xscale('log')
plt.yscale('log')
plt.title('Orbital Period vs Mass of Expolanets')
plt.xlabel('Orbital Period')
plt.ylabel('Mass')
plt.legend(title = 'Discovery Method')
plt.show()
```

```python
#2.2
planets_df = sns.load_dataset('planets')
planets_df = planets_df.dropna()

plt.figure(figsize = (15, 6))
sns.countplot(data = planets_df, x = 'year', hue = 'method')
plt.title('Discovery of Expolanets by Year and Method')
plt.xlabel('Year')
plt.ylabel('Amount of Exoplanets Discovered')
plt.legend(title = 'Discovery Method')
plt.show()
```
